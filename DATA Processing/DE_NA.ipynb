# 数据预处理的第二步：插补和删除空缺值

```python
# 导入pandas的包
import pandas as pd
import numpy as np
# 读入文件并 展示前五行数据
print("start")
# low_memory = False  就是读取列很多的数据的时候需要使用这个
data = pd.read_csv(r'F:\jupyter_notebook\Data\数据集\6组数据(带NA)\DO.csv',low_memory=False)
display(data.head(5))
print(data.shape)
print("over")

"""
删除所有带有NA的行，整体删除
"""

ft = data

sum = 0

for i,r in ft.iterrows():
    
    #print(r[-1],type(r[-1]))
    if str(r[-1]) == 'nan' :
        ft = ft.drop(index = i)
        print("========================================================")
        print("删除第%d行" %i)
        sum = sum +1
        
        
print("结束程序！共删除%d行" %sum)

#表示删除第一列
#ft.drop(ft.columns[0],axis = 1)
ft.to_csv(r"F:\jupyter_notebook\Data\数据集\6组数据(不带NA)\DO.csv",index=False)


print("保存完成！")
```
