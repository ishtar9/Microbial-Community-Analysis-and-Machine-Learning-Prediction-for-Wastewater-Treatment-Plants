# 数据预处理的第三步：异常值处理 箱线图法

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# 读取excel文件
data = pd.read_csv(r'F:\jupyter_notebook\Data\数据集\6组数据(不带NA)\DO.csv')
 
fig = plt.figure(figsize=(4, 4))

d = data['DO']


# 上四分位数
Q3 = d.quantile(0.75)
# 下四分位数
Q1 = d.quantile(0.25)

# 上限值
up_value = Q3 + 1.5 * (Q3 - Q1)
# 下限值
low_value = Q3 - 1.5 * (Q3 - Q1)


plt.boxplot([d], sym='*', boxprops={'color': 'orangered'})
plt.xticks(fontproperties='KaiTi')
plt.xticks(alpha=0)
plt.tick_params(axis='x', width=0)
plt.xlabel('DO', fontproperties='KaiTi', size=12)
plt.ylabel('value', fontproperties='KaiTi', size=12)
plt.savefig('DO.jpg', dpi=1000,bbox_inches='tight')
plt.show()



```

```python


ft = data

sum = 0

for i,r in ft.iterrows():
    
    #print(r[-1],type(r[-1]))
    if r[-1]>up_value or r[-1]<low_value:
        ft = ft.drop(index = i)
        print("========================================================")
        print("删除第%d行" %i)
        sum = sum +1
        
        
print("结束程序！共删除%d行" %sum)

#表示删除第一列
#ft.drop(ft.columns[0],axis = 1)
ft.to_csv(r"F:\jupyter_notebook\Data\数据集\6组数据(处理完异常值，不带NA)\DO.csv",index=False)

print("保存完成！")
```
